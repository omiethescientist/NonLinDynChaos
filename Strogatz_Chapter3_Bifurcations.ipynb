#  Saddle Node Bifurcation Example 
$ y(x) = r - x - e^{-x} $

```julia
f(x, p) = p - x
g(x) = exp(-x)

@manipulate for r = -5:0.1:5
    y(x) = r - x - exp(-x)
    plot(-5:0.1:5, x -> f(x, r); framestyle = :origin, ticks = false, label = "r-x", left_margin = 5Plots.mm, right_margin = 15Plots.mm)
    plot!(-2.5:.01:5, x -> g(x); label = "e^-x")
    roots = find_zeros(y, -5, 5)
    scatter!(roots, zeros(length(roots)), label = "x*")
    xlabel!("x")
    ylabel!("f(x)")
end
```

# Laser Threshold Example Phase Portrait

```julia
y(n, N₀, G, k, α) = (G*N₀ - k)*n - (α*G)*n^2
@manipulate for N₀ = 0:0.1:5, G = 0:0.1:5, k = 0:0.1:5, α = 0:0.1:5
    y(n) = (G*N₀ - k)*n - (α*G)*n^2
    plot(0:0.1:5, x -> y(x); framestyle = :origin, ticks = false, label = "N₀ = $(N₀); k/G = $(k/G)",left_margin = 5Plots.mm, right_margin = 15Plots.mm)
    roots = find_zeros(y, 0, 5)
    scatter!(roots, zeros(length(roots)), label = "x*")
    xlabel!("n")
    ylabel!("dn")
end
```

# Subcritical Pitchfork  With Stablizing Term (x^5) Bifurcation Diagram

```julia
y(u, r) = r*u + u^3 - u^5
xs = [find_zeros(x -> y(x, r), -5, 5) for r in -1:0.01:1]; 
```

```julia
rs = -1:0.01:1
c1 = [(rs[x],xs[x]...) for x in 1:length(xs) if length(xs[x]) == 1] 
c2 = [(rs[x],xs[x]...) for x in 1:length(xs) if length(xs[x]) == 3]
c3 = [(rs[x], xs[x]...) for x in 1:length(xs) if length(xs[x]) == 5];
```

```julia
c1 = reinterpret(reshape, Float64, c1)'
c2 = reinterpret(reshape, Float64, c2)'
c3 = reinterpret(reshape, Float64, c3)';
```

```julia
function plot_subcritical_fold(c)
    for i in 1:size(c, 2) -1
        if i% 2 > 0 
            plot!(c[:, 1], c[:, i+1],  framestyle = :origin, ticks = false, label = "Stable",
                left_margin = 5Plots.mm, right_margin = 15Plots.mm, linewidth = 2)
        else
            plot!(c[:, 1], c[:, i+1],  framestyle = :origin, ticks = false, label = "Unstable", linestyle = :dash, 
                left_margin = 5Plots.mm, right_margin = 15Plots.mm, linewidth = 2, legend = false)
        end
    end
end
```

```julia
plot()
plot_subcritical_fold(c1)
plot_subcritical_fold(c2)
plot_subcritical_fold(c3)
xlabel!("x*")
ylabel!("r")
```

# Imperfect Bifurcation with Sliding Imperfection Parameter

```julia
@manipulate  for h = -15:0.5:15 
    y(x, r) = h + r*x - x^3
    rs = -15:0.05:15
    xs = [find_zeros(x -> y(x, r), -20, 20) for r in -15:0.05:15] 
    c1 = [(rs[x],xs[x]...) for x in 1:length(xs) if length(xs[x]) == 1]
    #Not totally accurate does not represent the points where the number of roots is 2
    c2 = [(rs[x],xs[x]...) for x in 1:length(xs) if length(xs[x]) == 3]
    c1 = reinterpret(reshape, Float64, c1)'
    c2 = reinterpret(reshape, Float64, c2)'
    plot()
    plot_subcritical_fold(c1)
    plot_subcritical_fold(c2)
    xlabel!("x*")
    ylabel!("r")
end
```

# Insect Outbreak Example: Intersections

```julia
@manipulate for r = 0:0.5:15, k = 0:0.5:16 
    y(x) = r*x*(1-x/k) - x^2/(1+x^2)
    f(x) = r*(1-x/k)
    g(x) = x/(1+x^2)
    plot(0:0.1:5, x -> f(x); framestyle = :origin, ticks = false, label = "y = r(1-x/k)",left_margin = 5Plots.mm, right_margin = 15Plots.mm)
    plot!(0:0.1:5, x -> g(x), label = "y = x/(1+x^2)")
    roots = find_zeros(y, -5, 5)
    scatter!(roots, zeros(length(roots)), label = "x*")
    xlabel!("x")
    ylabel!("f(x)")
end
```
